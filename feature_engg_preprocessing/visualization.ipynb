# Matplotlib


 <h3 style="text-align:center"> Architecture </h3>
  
   Layers of Matplotlib : <br>
     Scripting layer(pyplot) ->Artist Layer (Artist) ->Backend Layer (Figure Canvas,Renderer,Event )

## Backend  Layer
     
          
       1. Figure canvas :Encompasses the area to draw the picture 
       2. Rendere       :know how to draw on the figure canvas
       3. Event         : Handles user input such as key strokes and mouse clicks
       
## Artist layer
      1. it comprises of one main object Artist
      2. It knows how to use rendrer to draw on canvas
      3. Title ,line ,lables all crrospond to artist instance
      
### there are two types of artist object 
      1. primitive :Line2d ,rectangle and circle
      2. Composite : Axes ,tick ,axis ,figure
