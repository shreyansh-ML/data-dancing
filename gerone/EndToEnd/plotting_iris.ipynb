```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df1=pd.DataFrame(data=iris.data,columns= ['sl','sw','pl','pw'])
```

```python
df2=pd.DataFrame(data=iris.target,columns=['label'])
```

```python
df3=pd.concat([df1,df2],axis=1)
```

```python
df3.head()
```

```python
plt.scatter(df3['sl'],df3['sw'],c=df3['label'],cmap=plt.get_cmap("jet"))
plt.legend()
```
